```python
# Importar bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Criar o DataFrame com os dados fornecidos
data = {
    'Dias': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Horas_Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs_Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas_Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(data)
```

```python
# Total de Horas Trabalhadas
total_horas_trabalhadas = df['Horas_Trabalhadas'].sum()
print(f'Total de Horas Trabalhadas: {total_horas_trabalhadas} horas')
```

```python
# Média Diária de Horas Trabalhadas
media_diaria_horas = df['Horas_Trabalhadas'].mean()
print(f'Média Diária de Horas Trabalhadas: {media_diaria_horas:.2f} horas')
```

```python
# Total de bugs corrigidos
total_bugs_corrigidos = df['Bugs_Corrigidos'].sum()
print(f'Total de Bugs Corrigidos: {total_bugs_corrigidos} bugs')
```

```python
# Média diária de bugs corrigidos
media_diaria_bugs_corrigidos = total_bugs_corrigidos / len(data)
print("Média diária de bugs corrigidos:", media_diaria_bugs_corrigidos)
```

```python
# Total de tarefas concluídas
total_tarefas_concluidas = df['Tarefas_Concluídas'].sum()
print(f'Total de Tarefas Concluídas: {total_tarefas_concluidas} tarefas')
```

```python
# Média diária de tarefas concluídas
media_diaria_tarefas = df['Tarefas_Concluídas'].mean()
print(f'Média Diária de Tarefas Concluídas: {media_diaria_tarefas:.2f} tarefas')
```

```python
# Produtividade diária (tarefas concluídas por hora)
df['Produtividade'] = df['Tarefas_Concluídas'] / df['Horas_Trabalhadas']
print(df[['Dias', 'Produtividade']])
```

```python
# Gráfico de Barras
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
sns.barplot(x='Dias', y='Horas_Trabalhadas', data=df, palette='viridis')
plt.title('Horas Trabalhadas por Dia')

plt.subplot(3, 1, 2)
sns.barplot(x='Dias', y='Bugs_Corrigidos', data=df, palette='viridis')
plt.title('Bugs Corrigidos por Dia')

plt.subplot(3, 1, 3)
sns.barplot(x='Dias', y='Tarefas_Concluídas', data=df, palette='viridis')
plt.title('Tarefas Concluídas por Dia')

plt.tight_layout()
plt.show()
```

```python
# Gráfico de Linhas para Produtividade Diária
plt.figure(figsize=(10, 6))
sns.lineplot(x='Dias', y='Produtividade', data=df, marker='o', color='purple')
plt.title('Produtividade Diária (Tarefas Concluídas por Hora)')
plt.xlabel('Dias')
plt.ylabel('Produtividade')
plt.show()
```
